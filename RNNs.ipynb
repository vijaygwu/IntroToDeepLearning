```python
import torch
import torch.nn as nn
from torch.nn.utils.rnn import pad_sequence, pack_padded_sequence, pad_packed_sequence
from torch.utils.data import Dataset, DataLoader
from torch.optim import Adam
from tqdm import tqdm
import matplotlib.pyplot as plt
import time
import random
import re
import requests
from io import BytesIO
import tarfile
import os

# Check if CUDA is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Download and extract the IMDB dataset
url = "https://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz"
response = requests.get(url)
file = tarfile.open(fileobj=BytesIO(response.content), mode="r:gz")
file.extractall()
file.close()

# Data Preparation
def load_imdb_data(split):
    data = []
    for sentiment in ['pos', 'neg']:
        path = f'aclImdb/{split}/{sentiment}'
        for filename in os.listdir(path):
            with open(os.path.join(path, filename), 'r', encoding='utf-8') as file:
                review = file.read()
                data.append((review, 1 if sentiment == 'pos' else 0))
    return data

def preprocess_text(text):
    text = re.sub(r'<[^>]+>', '', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = text.lower().strip()
    return text.split()

# Custom Dataset
class IMDBDataset(Dataset):
    def __init__(self, data, vocab):
        self.data = data
        self.vocab = vocab

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        text, label = self.data[idx]
        tokens = preprocess_text(text)
        return torch.tensor([self.vocab.get(token, self.vocab['<unk>']) for token in tokens]), label

# Collate function for padding
def collate_fn(batch):
    texts, labels = zip(*batch)
    lengths = torch.LongTensor([len(text) for text in texts])
    padded_texts = pad_sequence(texts, batch_first=True, padding_value=0)
    return padded_texts, lengths, torch.LongTensor(labels)

# Model definitions
class SimpleRNN(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_size, output_size):
        super(SimpleRNN, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.rnn = nn.RNN(embedding_dim, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        embedded = self.embedding(x)
        _, hidden = self.rnn(embedded)
        return self.fc(hidden.squeeze(0))

class PackedRNN(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_size, output_size):
        super(PackedRNN, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.rnn = nn.RNN(embedding_dim, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x, lengths):
        embedded = self.embedding(x)
        packed = pack_padded_sequence(embedded, lengths.cpu(), batch_first=True, enforce_sorted=False)
        _, hidden = self.rnn(packed)
        return self.fc(hidden.squeeze(0))

class BidirectionalRNN(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_size, output_size):
        super(BidirectionalRNN, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.rnn = nn.RNN(embedding_dim, hidden_size, batch_first=True, bidirectional=True)
        self.fc = nn.Linear(hidden_size * 2, output_size)

    def forward(self, x, lengths):
        embedded = self.embedding(x)
        packed = pack_padded_sequence(embedded, lengths.cpu(), batch_first=True, enforce_sorted=False)
        _, hidden = self.rnn(packed)
        hidden = torch.cat((hidden[-2,:,:], hidden[-1,:,:]), dim=1)
        return self.fc(hidden)

# Training Function
def train(model, dataloader, optimizer, criterion, device):
    model.train()
    total_loss = 0
    for batch in tqdm(dataloader, desc="Training", leave=False):
        x, lengths, y = batch
        x, y = x.to(device), y.to(device)
        lengths = lengths.to(device)

        optimizer.zero_grad()
        if isinstance(model, SimpleRNN):
            output = model(x)
        else:
            output = model(x, lengths)
        loss = criterion(output, y)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
    return total_loss / len(dataloader)

# Evaluation Function
def evaluate(model, dataloader, criterion, device):
    model.eval()
    total_loss = 0
    correct = 0
    with torch.no_grad():
        for batch in tqdm(dataloader, desc="Evaluating", leave=False):
            x, lengths, y = batch
            x, y = x.to(device), y.to(device)
            lengths = lengths.to(device)

            if isinstance(model, SimpleRNN):
                output = model(x)
            else:
                output = model(x, lengths)
            loss = criterion(output, y)
            total_loss += loss.item()

            pred = output.argmax(dim=1, keepdim=True)
            correct += pred.eq(y.view_as(pred)).sum().item()

    accuracy = correct / len(dataloader.dataset)
    return total_loss / len(dataloader), accuracy

def visualize_results(accuracies, times):
    models = list(accuracies.keys())

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # Accuracy plot
    for model in models:
        ax1.plot(range(1, len(accuracies[model])+1), accuracies[model], label=model)
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Accuracy')
    ax1.set_title('Test Accuracy vs Epoch')
    ax1.legend()

    # Time plot
    ax2.bar(models, times.values())
    ax2.set_ylabel('Time (seconds)')
    ax2.set_title('Total Training Time')
    ax2.tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.show()

# Main Execution
if __name__ == "__main__":
    # Load and preprocess data
    train_data = load_imdb_data('train')
    test_data = load_imdb_data('test')

    # Create vocabulary
    vocab = {'<pad>': 0, '<unk>': 1}
    for text, _ in train_data:
        for token in preprocess_text(text):
            if token not in vocab:
                vocab[token] = len(vocab)

    # Create datasets and dataloaders
    train_dataset = IMDBDataset(train_data, vocab)
    test_dataset = IMDBDataset(test_data, vocab)
    train_dataloader = DataLoader(train_dataset, batch_size=64, collate_fn=collate_fn, shuffle=True)
    test_dataloader = DataLoader(test_dataset, batch_size=64, collate_fn=collate_fn)

    # Hyperparameters
    EMBEDDING_DIM = 100
    HIDDEN_SIZE = 256
    OUTPUT_SIZE = 2
    EPOCHS = 100
    LEARNING_RATE = 0.001

    # Initialize models
    simple_rnn = SimpleRNN(len(vocab), EMBEDDING_DIM, HIDDEN_SIZE, OUTPUT_SIZE).to(device)
    packed_rnn = PackedRNN(len(vocab), EMBEDDING_DIM, HIDDEN_SIZE, OUTPUT_SIZE).to(device)
    bidirectional_rnn = BidirectionalRNN(len(vocab), EMBEDDING_DIM, HIDDEN_SIZE, OUTPUT_SIZE).to(device)

    # Dictionaries to store results
    accuracies = {"Simple RNN": [], "Packed RNN": [], "Bidirectional RNN": []}
    times = {"Simple RNN": 0, "Packed RNN": 0, "Bidirectional RNN": 0}

    # Training loop
    for model_name, model in [("Simple RNN", simple_rnn), ("Packed RNN", packed_rnn), ("Bidirectional RNN", bidirectional_rnn)]:
        print(f"\nTraining {model_name}")
        optimizer = Adam(model.parameters(), lr=LEARNING_RATE)
        criterion = nn.CrossEntropyLoss()

        start_time = time.time()
        for epoch in range(EPOCHS):
            train_loss = train(model, train_dataloader, optimizer, criterion, device)
            test_loss, test_acc = evaluate(model, test_dataloader, criterion, device)
            accuracies[model_name].append(test_acc)
            print(f"Epoch {epoch+1}/{EPOCHS}, Train Loss: {train_loss:.4f}, Test Loss: {test_loss:.4f}, Test Acc: {test_acc:.4f}")

        end_time = time.time()
        times[model_name] = end_time - start_time
        print(f"Total training time for {model_name}: {times[model_name]:.2f} seconds")

    # Visualize results
    visualize_results(accuracies, times)
```

```python
import torch
import torch.nn as nn
from torch.nn.utils.rnn import pad_sequence, pack_padded_sequence, pad_packed_sequence
from torch.utils.data import Dataset, DataLoader
from torch.optim import Adam
from torch.optim.lr_scheduler import ReduceLROnPlateau
from tqdm import tqdm
import matplotlib.pyplot as plt
import time
import random
import re
import requests
from io import BytesIO
import tarfile
import os

# Check if CUDA is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Download and extract the IMDB dataset
url = "https://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz"
response = requests.get(url)
file = tarfile.open(fileobj=BytesIO(response.content), mode="r:gz")
file.extractall()
file.close()

# Data Preparation
def load_imdb_data(split):
    data = []
    for sentiment in ['pos', 'neg']:
        path = f'aclImdb/{split}/{sentiment}'
        for filename in os.listdir(path):
            with open(os.path.join(path, filename), 'r', encoding='utf-8') as file:
                review = file.read()
                data.append((review, 1 if sentiment == 'pos' else 0))
    return data

def preprocess_text(text):
    text = re.sub(r'<[^>]+>', '', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = text.lower().strip()
    return text.split()

# Custom Dataset
class IMDBDataset(Dataset):
    def __init__(self, data, vocab):
        self.data = data
        self.vocab = vocab

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        text, label = self.data[idx]
        tokens = preprocess_text(text)
        return torch.tensor([self.vocab.get(token, self.vocab['<unk>']) for token in tokens]), label

# Collate function for padding
def collate_fn(batch):
    texts, labels = zip(*batch)
    lengths = torch.LongTensor([len(text) for text in texts])
    padded_texts = pad_sequence(texts, batch_first=True, padding_value=0)
    return padded_texts, lengths, torch.LongTensor(labels)

# Updated Model definitions with dropout
class SimpleRNN(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_size, output_size, dropout_rate=0.5):
        super(SimpleRNN, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.rnn = nn.RNN(embedding_dim, hidden_size, batch_first=True, dropout=dropout_rate)
        self.dropout = nn.Dropout(dropout_rate)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        embedded = self.embedding(x)
        _, hidden = self.rnn(embedded)
        dropped = self.dropout(hidden.squeeze(0))
        return self.fc(dropped)

class PackedRNN(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_size, output_size, dropout_rate=0.5):
        super(PackedRNN, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.rnn = nn.RNN(embedding_dim, hidden_size, batch_first=True, dropout=dropout_rate)
        self.dropout = nn.Dropout(dropout_rate)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x, lengths):
        embedded = self.embedding(x)
        packed = pack_padded_sequence(embedded, lengths.cpu(), batch_first=True, enforce_sorted=False)
        _, hidden = self.rnn(packed)
        dropped = self.dropout(hidden.squeeze(0))
        return self.fc(dropped)

class BidirectionalRNN(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_size, output_size, dropout_rate=0.5):
        super(BidirectionalRNN, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.rnn = nn.RNN(embedding_dim, hidden_size, batch_first=True, bidirectional=True, dropout=dropout_rate)
        self.dropout = nn.Dropout(dropout_rate)
        self.fc = nn.Linear(hidden_size * 2, output_size)

    def forward(self, x, lengths):
        embedded = self.embedding(x)
        packed = pack_padded_sequence(embedded, lengths.cpu(), batch_first=True, enforce_sorted=False)
        _, hidden = self.rnn(packed)
        hidden = torch.cat((hidden[-2,:,:], hidden[-1,:,:]), dim=1)
        dropped = self.dropout(hidden)
        return self.fc(dropped)

# Training Function
def train(model, dataloader, optimizer, criterion, device):
    model.train()
    total_loss = 0
    for batch in tqdm(dataloader, desc="Training", leave=False):
        x, lengths, y = batch
        x, y = x.to(device), y.to(device)
        lengths = lengths.to(device)

        optimizer.zero_grad()
        if isinstance(model, SimpleRNN):
            output = model(x)
        else:
            output = model(x, lengths)
        loss = criterion(output, y)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
    return total_loss / len(dataloader)

# Evaluation Function
def evaluate(model, dataloader, criterion, device):
    model.eval()
    total_loss = 0
    correct = 0
    with torch.no_grad():
        for batch in tqdm(dataloader, desc="Evaluating", leave=False):
            x, lengths, y = batch
            x, y = x.to(device), y.to(device)
            lengths = lengths.to(device)

            if isinstance(model, SimpleRNN):
                output = model(x)
            else:
                output = model(x, lengths)
            loss = criterion(output, y)
            total_loss += loss.item()

            pred = output.argmax(dim=1, keepdim=True)
            correct += pred.eq(y.view_as(pred)).sum().item()

    accuracy = correct / len(dataloader.dataset)
    return total_loss / len(dataloader), accuracy

def visualize_results(accuracies, times, learning_rates):
    models = list(accuracies.keys())

    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 5))

    # Accuracy plot
    for model in models:
        ax1.plot(range(1, len(accuracies[model])+1), accuracies[model], label=model)
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Accuracy')
    ax1.set_title('Test Accuracy vs Epoch')
    ax1.legend()

    # Time plot
    ax2.bar(models, times.values())
    ax2.set_ylabel('Time (seconds)')
    ax2.set_title('Total Training Time')
    ax2.tick_params(axis='x', rotation=45)

    # Learning rate plot
    for model in models:
        ax3.plot(range(1, len(learning_rates[model])+1), learning_rates[model], label=model)
    ax3.set_xlabel('Epoch')
    ax3.set_ylabel('Learning Rate')
    ax3.set_title('Learning Rate vs Epoch')
    ax3.legend()

    plt.tight_layout()
    plt.show()

# Main Execution
if __name__ == "__main__":
    # Load and preprocess data
    train_data = load_imdb_data('train')
    test_data = load_imdb_data('test')

    # Create vocabulary
    vocab = {'<pad>': 0, '<unk>': 1}
    for text, _ in train_data:
        for token in preprocess_text(text):
            if token not in vocab:
                vocab[token] = len(vocab)

    # Create datasets and dataloaders
    train_dataset = IMDBDataset(train_data, vocab)
    test_dataset = IMDBDataset(test_data, vocab)

    # Increased batch size
    BATCH_SIZE = 256
    train_dataloader = DataLoader(train_dataset, batch_size=BATCH_SIZE, collate_fn=collate_fn, shuffle=True)
    test_dataloader = DataLoader(test_dataset, batch_size=BATCH_SIZE, collate_fn=collate_fn)

    # Hyperparameters
    EMBEDDING_DIM = 100
    HIDDEN_SIZE = 256
    OUTPUT_SIZE = 2
    EPOCHS = 10
    LEARNING_RATE = 0.001
    DROPOUT_RATE = 0.5

    # Initialize models
    simple_rnn = SimpleRNN(len(vocab), EMBEDDING_DIM, HIDDEN_SIZE, OUTPUT_SIZE, DROPOUT_RATE).to(device)
    packed_rnn = PackedRNN(len(vocab), EMBEDDING_DIM, HIDDEN_SIZE, OUTPUT_SIZE, DROPOUT_RATE).to(device)
    bidirectional_rnn = BidirectionalRNN(len(vocab), EMBEDDING_DIM, HIDDEN_SIZE, OUTPUT_SIZE, DROPOUT_RATE).to(device)

    # Dictionaries to store results
    accuracies = {"Simple RNN": [], "Packed RNN": [], "Bidirectional RNN": []}
    times = {"Simple RNN": 0, "Packed RNN": 0, "Bidirectional RNN": 0}
    learning_rates = {"Simple RNN": [], "Packed RNN": [], "Bidirectional RNN": []}

    # Training loop
    for model_name, model in [("Simple RNN", simple_rnn), ("Packed RNN", packed_rnn), ("Bidirectional RNN", bidirectional_rnn)]:
        print(f"\nTraining {model_name}")
        optimizer = Adam(model.parameters(), lr=LEARNING_RATE)
        scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=1, verbose=True)
        criterion = nn.CrossEntropyLoss()

        start_time = time.time()
        for epoch in range(EPOCHS):
            train_loss = train(model, train_dataloader, optimizer, criterion, device)
            test_loss, test_acc = evaluate(model, test_dataloader, criterion, device)
            accuracies[model_name].append(test_acc)
            learning_rates[model_name].append(optimizer.param_groups[0]['lr'])

            print(f"Epoch {epoch+1}/{EPOCHS}, Train Loss: {train_loss:.4f}, Test Loss: {test_loss:.4f}, Test Acc: {test_acc:.4f}, LR: {optimizer.param_groups[0]['lr']:.6f}")

            # Step the scheduler
            scheduler.step(test_loss)

        end_time = time.time()
        times[model_name] = end_time - start_time
        print(f"Total training time for {model_name}: {times[model_name]:.2f} seconds")

    # Visualize results
    visualize_results(accuracies, times, learning_rates)
```

```python
import torch
import torch.nn as nn
from torch.nn.utils.rnn import pad_sequence, pack_padded_sequence, pad_packed_sequence
from torch.utils.data import Dataset, DataLoader
from torch.optim import Adam
from torch.optim.lr_scheduler import ReduceLROnPlateau
from tqdm import tqdm
import matplotlib.pyplot as plt
import time
import random
import re
import requests
from io import BytesIO
import tarfile
import os

# Check if CUDA is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Download and extract the IMDB dataset
url = "https://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz"
response = requests.get(url)
file = tarfile.open(fileobj=BytesIO(response.content), mode="r:gz")
file.extractall()
file.close()

# Data Preparation
def load_imdb_data(split):
    data = []
    for sentiment in ['pos', 'neg']:
        path = f'aclImdb/{split}/{sentiment}'
        for filename in os.listdir(path):
            with open(os.path.join(path, filename), 'r', encoding='utf-8') as file:
                review = file.read()
                data.append((review, 1 if sentiment == 'pos' else 0))
    return data

def preprocess_text(text):
    text = re.sub(r'<[^>]+>', '', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = text.lower().strip()
    return text.split()

# Custom Dataset
class IMDBDataset(Dataset):
    def __init__(self, data, vocab):
        self.data = data
        self.vocab = vocab

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        text, label = self.data[idx]
        tokens = preprocess_text(text)
        return torch.tensor([self.vocab.get(token, self.vocab['<unk>']) for token in tokens]), label

# Collate function for padding
def collate_fn(batch):
    texts, labels = zip(*batch)
    lengths = torch.LongTensor([len(text) for text in texts])
    padded_texts = pad_sequence(texts, batch_first=True, padding_value=0)
    return padded_texts, lengths, torch.LongTensor(labels)

# Model definitions with dropout
class SimpleRNN(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_size, output_size, dropout_rate=0.5):
        super(SimpleRNN, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.rnn = nn.RNN(embedding_dim, hidden_size, batch_first=True, dropout=dropout_rate)
        self.dropout = nn.Dropout(dropout_rate)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        embedded = self.embedding(x)
        _, hidden = self.rnn(embedded)
        dropped = self.dropout(hidden.squeeze(0))
        return self.fc(dropped)

class PackedRNN(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_size, output_size, dropout_rate=0.5):
        super(PackedRNN, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.rnn = nn.RNN(embedding_dim, hidden_size, batch_first=True, dropout=dropout_rate)
        self.dropout = nn.Dropout(dropout_rate)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x, lengths):
        embedded = self.embedding(x)
        packed = pack_padded_sequence(embedded, lengths.cpu(), batch_first=True, enforce_sorted=False)
        _, hidden = self.rnn(packed)
        dropped = self.dropout(hidden.squeeze(0))
        return self.fc(dropped)

class BidirectionalRNN(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_size, output_size, dropout_rate=0.5):
        super(BidirectionalRNN, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.rnn = nn.RNN(embedding_dim, hidden_size, batch_first=True, bidirectional=True, dropout=dropout_rate)
        self.dropout = nn.Dropout(dropout_rate)
        self.fc = nn.Linear(hidden_size * 2, output_size)

    def forward(self, x, lengths):
        embedded = self.embedding(x)
        packed = pack_padded_sequence(embedded, lengths.cpu(), batch_first=True, enforce_sorted=False)
        _, hidden = self.rnn(packed)
        hidden = torch.cat((hidden[-2,:,:], hidden[-1,:,:]), dim=1)
        dropped = self.dropout(hidden)
        return self.fc(dropped)

# Training Function with Gradient Clipping
def train(model, dataloader, optimizer, criterion, device, l2_lambda, clip_value):
    model.train()
    total_loss = 0
    for batch in tqdm(dataloader, desc="Training", leave=False):
        x, lengths, y = batch
        x, y = x.to(device), y.to(device)
        lengths = lengths.to(device)

        optimizer.zero_grad()
        if isinstance(model, SimpleRNN):
            output = model(x)
        else:
            output = model(x, lengths)

        # Calculate L2 regularization term
        l2_reg = torch.tensor(0.).to(device)
        for param in model.parameters():
            l2_reg += torch.norm(param)

        loss = criterion(output, y) + l2_lambda * l2_reg
        loss.backward()

        # Gradient clipping
        torch.nn.utils.clip_grad_norm_(model.parameters(), clip_value)

        optimizer.step()

        total_loss += loss.item()
    return total_loss / len(dataloader)

# Evaluation Function
def evaluate(model, dataloader, criterion, device):
    model.eval()
    total_loss = 0
    correct = 0
    with torch.no_grad():
        for batch in tqdm(dataloader, desc="Evaluating", leave=False):
            x, lengths, y = batch
            x, y = x.to(device), y.to(device)
            lengths = lengths.to(device)

            if isinstance(model, SimpleRNN):
                output = model(x)
            else:
                output = model(x, lengths)
            loss = criterion(output, y)
            total_loss += loss.item()

            pred = output.argmax(dim=1, keepdim=True)
            correct += pred.eq(y.view_as(pred)).sum().item()

    accuracy = correct / len(dataloader.dataset)
    return total_loss / len(dataloader), accuracy

def visualize_results(accuracies, times, learning_rates):
    models = list(accuracies.keys())

    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 5))

    # Accuracy plot
    for model in models:
        ax1.plot(range(1, len(accuracies[model])+1), accuracies[model], label=model)
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Accuracy')
    ax1.set_title('Test Accuracy vs Epoch')
    ax1.legend()

    # Time plot
    ax2.bar(models, times.values())
    ax2.set_ylabel('Time (seconds)')
    ax2.set_title('Total Training Time')
    ax2.tick_params(axis='x', rotation=45)

    # Learning rate plot
    for model in models:
        ax3.plot(range(1, len(learning_rates[model])+1), learning_rates[model], label=model)
    ax3.set_xlabel('Epoch')
    ax3.set_ylabel('Learning Rate')
    ax3.set_title('Learning Rate vs Epoch')
    ax3.legend()

    plt.tight_layout()
    plt.show()

# Main Execution
if __name__ == "__main__":
    # Load and preprocess data
    train_data = load_imdb_data('train')
    test_data = load_imdb_data('test')

    # Create vocabulary
    vocab = {'<pad>': 0, '<unk>': 1}
    for text, _ in train_data:
        for token in preprocess_text(text):
            if token not in vocab:
                vocab[token] = len(vocab)

    # Create datasets and dataloaders
    train_dataset = IMDBDataset(train_data, vocab)
    test_dataset = IMDBDataset(test_data, vocab)

    # Increased batch size
    BATCH_SIZE = 256
    train_dataloader = DataLoader(train_dataset, batch_size=BATCH_SIZE, collate_fn=collate_fn, shuffle=True)
    test_dataloader = DataLoader(test_dataset, batch_size=BATCH_SIZE, collate_fn=collate_fn)

    # Hyperparameters
    EMBEDDING_DIM = 100
    HIDDEN_SIZE = 256
    OUTPUT_SIZE = 2
    EPOCHS = 10
    LEARNING_RATE = 0.001
    DROPOUT_RATE = 0.5
    L2_LAMBDA = 1e-5  # L2 regularization strength
    CLIP_VALUE = 1.0  # Gradient clipping threshold

    # Initialize models
    simple_rnn = SimpleRNN(len(vocab), EMBEDDING_DIM, HIDDEN_SIZE, OUTPUT_SIZE, DROPOUT_RATE).to(device)
    packed_rnn = PackedRNN(len(vocab), EMBEDDING_DIM, HIDDEN_SIZE, OUTPUT_SIZE, DROPOUT_RATE).to(device)
    bidirectional_rnn = BidirectionalRNN(len(vocab), EMBEDDING_DIM, HIDDEN_SIZE, OUTPUT_SIZE, DROPOUT_RATE).to(device)

    # Dictionaries to store results
    accuracies = {"Simple RNN": [], "Packed RNN": [], "Bidirectional RNN": []}
    times = {"Simple RNN": 0, "Packed RNN": 0, "Bidirectional RNN": 0}
    learning_rates = {"Simple RNN": [], "Packed RNN": [], "Bidirectional RNN": []}

    # Training loop
    for model_name, model in [("Simple RNN", simple_rnn), ("Packed RNN", packed_rnn), ("Bidirectional RNN", bidirectional_rnn)]:
        print(f"\nTraining {model_name}")
        optimizer = Adam(model.parameters(), lr=LEARNING_RATE, weight_decay=L2_LAMBDA)
        scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=1, verbose=True)
        criterion = nn.CrossEntropyLoss()

        start_time = time.time()
        for epoch in range(EPOCHS):
            train_loss = train(model, train_dataloader, optimizer, criterion, device, L2_LAMBDA, CLIP_VALUE)
            test_loss, test_acc = evaluate(model, test_dataloader, criterion, device)
            accuracies[model_name].append(test_acc)
            learning_rates[model_name].append(optimizer.param_groups[0]['lr'])

            print(f"Epoch {epoch+1}/{EPOCHS}, Train Loss: {train_loss:.4f}, Test Loss: {test_loss:.4f}, Test Acc: {test_acc:.4f}, LR: {optimizer.param_groups[0]['lr']:.6f}")

            # Step the scheduler
            scheduler.step(test_loss)

        end_time = time.time()
        times[model_name] = end_time - start_time
        print(f"Total training time for {model_name}: {times[model_name]:.2f} seconds")

    # Visualize results
    visualize_results(accuracies, times, learning_rates)
```

```python
import torch
import torch.nn as nn
from torch.nn.utils.rnn import pad_sequence, pack_padded_sequence, pad_packed_sequence
from torch.utils.data import Dataset, DataLoader
from torch.optim import Adam
from torch.optim.lr_scheduler import ReduceLROnPlateau
from tqdm import tqdm
import matplotlib.pyplot as plt
import time
import random
import re
import requests
from io import BytesIO
import tarfile
import os

# Check if CUDA is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Download and extract the IMDB dataset
url = "https://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz"
response = requests.get(url)
file = tarfile.open(fileobj=BytesIO(response.content), mode="r:gz")
file.extractall()
file.close()

# Data Preparation
def load_imdb_data(split):
    data = []
    for sentiment in ['pos', 'neg']:
        path = f'aclImdb/{split}/{sentiment}'
        for filename in os.listdir(path):
            with open(os.path.join(path, filename), 'r', encoding='utf-8') as file:
                review = file.read()
                data.append((review, 1 if sentiment == 'pos' else 0))
    return data

def preprocess_text(text):
    text = re.sub(r'<[^>]+>', '', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = text.lower().strip()
    return text.split()

# Custom Dataset
class IMDBDataset(Dataset):
    def __init__(self, data, vocab):
        self.data = data
        self.vocab = vocab

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        text, label = self.data[idx]
        tokens = preprocess_text(text)
        return torch.tensor([self.vocab.get(token, self.vocab['<unk>']) for token in tokens]), label

# Collate function for padding
def collate_fn(batch):
    texts, labels = zip(*batch)
    lengths = torch.LongTensor([len(text) for text in texts])
    padded_texts = pad_sequence(texts, batch_first=True, padding_value=0)
    return padded_texts, lengths, torch.LongTensor(labels)

# Model definitions
class SimpleRNN(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_size, output_size, dropout_rate=0.5):
        super(SimpleRNN, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.rnn = nn.RNN(embedding_dim, hidden_size, batch_first=True, dropout=dropout_rate)
        self.dropout = nn.Dropout(dropout_rate)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        embedded = self.embedding(x)
        _, hidden = self.rnn(embedded)
        dropped = self.dropout(hidden.squeeze(0))
        return self.fc(dropped)

class PackedRNN(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_size, output_size, dropout_rate=0.5):
        super(PackedRNN, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.rnn = nn.RNN(embedding_dim, hidden_size, batch_first=True, dropout=dropout_rate)
        self.dropout = nn.Dropout(dropout_rate)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x, lengths):
        embedded = self.embedding(x)
        packed = pack_padded_sequence(embedded, lengths.cpu(), batch_first=True, enforce_sorted=False)
        _, hidden = self.rnn(packed)
        dropped = self.dropout(hidden.squeeze(0))
        return self.fc(dropped)

class BidirectionalRNN(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_size, output_size, dropout_rate=0.5):
        super(BidirectionalRNN, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.rnn = nn.RNN(embedding_dim, hidden_size, batch_first=True, bidirectional=True, dropout=dropout_rate)
        self.dropout = nn.Dropout(dropout_rate)
        self.fc = nn.Linear(hidden_size * 2, output_size)

    def forward(self, x, lengths):
        embedded = self.embedding(x)
        packed = pack_padded_sequence(embedded, lengths.cpu(), batch_first=True, enforce_sorted=False)
        _, hidden = self.rnn(packed)
        hidden = torch.cat((hidden[-2,:,:], hidden[-1,:,:]), dim=1)
        dropped = self.dropout(hidden)
        return self.fc(dropped)

class GRUModel(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_size, output_size, n_layers, dropout_rate=0.5):
        super(GRUModel, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.gru = nn.GRU(embedding_dim, hidden_size, n_layers, batch_first=True, dropout=dropout_rate if n_layers > 1 else 0, bidirectional=True)
        self.fc = nn.Linear(hidden_size * 2, output_size)
        self.dropout = nn.Dropout(dropout_rate)

    def forward(self, x, lengths):
        embedded = self.embedding(x)
        packed = pack_padded_sequence(embedded, lengths.cpu(), batch_first=True, enforce_sorted=False)
        _, hidden = self.gru(packed)
        hidden = self.dropout(torch.cat((hidden[-2,:,:], hidden[-1,:,:]), dim=1))
        return self.fc(hidden)

# Training Function with Gradient Clipping
def train(model, dataloader, optimizer, criterion, device, l2_lambda, clip_value):
    model.train()
    total_loss = 0
    for batch in tqdm(dataloader, desc="Training", leave=False):
        x, lengths, y = batch
        x, y = x.to(device), y.to(device)
        lengths = lengths.to(device)

        optimizer.zero_grad()
        if isinstance(model, SimpleRNN):
            output = model(x)
        else:
            output = model(x, lengths)

        # Calculate L2 regularization term
        l2_reg = torch.tensor(0.).to(device)
        for param in model.parameters():
            l2_reg += torch.norm(param)

        loss = criterion(output, y) + l2_lambda * l2_reg
        loss.backward()

        # Gradient clipping
        torch.nn.utils.clip_grad_norm_(model.parameters(), clip_value)

        optimizer.step()

        total_loss += loss.item()
    return total_loss / len(dataloader)

# Evaluation Function
def evaluate(model, dataloader, criterion, device):
    model.eval()
    total_loss = 0
    correct = 0
    with torch.no_grad():
        for batch in tqdm(dataloader, desc="Evaluating", leave=False):
            x, lengths, y = batch
            x, y = x.to(device), y.to(device)
            lengths = lengths.to(device)

            if isinstance(model, SimpleRNN):
                output = model(x)
            else:
                output = model(x, lengths)
            loss = criterion(output, y)
            total_loss += loss.item()

            pred = output.argmax(dim=1, keepdim=True)
            correct += pred.eq(y.view_as(pred)).sum().item()

    accuracy = correct / len(dataloader.dataset)
    return total_loss / len(dataloader), accuracy

def visualize_results(accuracies, times, learning_rates):
    models = list(accuracies.keys())

    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 5))

    # Accuracy plot
    for model in models:
        ax1.plot(range(1, len(accuracies[model])+1), accuracies[model], label=model)
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Accuracy')
    ax1.set_title('Test Accuracy vs Epoch')
    ax1.legend()

    # Time plot
    ax2.bar(models, times.values())
    ax2.set_ylabel('Time (seconds)')
    ax2.set_title('Total Training Time')
    ax2.tick_params(axis='x', rotation=45)

    # Learning rate plot
    for model in models:
        ax3.plot(range(1, len(learning_rates[model])+1), learning_rates[model], label=model)
    ax3.set_xlabel('Epoch')
    ax3.set_ylabel('Learning Rate')
    ax3.set_title('Learning Rate vs Epoch')
    ax3.legend()

    plt.tight_layout()
    plt.show()

# Main Execution
if __name__ == "__main__":
    # Load and preprocess data
    train_data = load_imdb_data('train')
    test_data = load_imdb_data('test')

    # Create vocabulary
    vocab = {'<pad>': 0, '<unk>': 1}
    for text, _ in train_data:
        for token in preprocess_text(text):
            if token not in vocab:
                vocab[token] = len(vocab)

    # Create datasets and dataloaders
    train_dataset = IMDBDataset(train_data, vocab)
    test_dataset = IMDBDataset(test_data, vocab)

    BATCH_SIZE = 256
    train_dataloader = DataLoader(train_dataset, batch_size=BATCH_SIZE, collate_fn=collate_fn, shuffle=True)
    test_dataloader = DataLoader(test_dataset, batch_size=BATCH_SIZE, collate_fn=collate_fn)

    # Hyperparameters
    EMBEDDING_DIM = 100
    HIDDEN_SIZE = 256
    OUTPUT_SIZE = 2
    N_LAYERS = 2
    EPOCHS = 10
    LEARNING_RATE = 0.001
    DROPOUT_RATE = 0.5
    L2_LAMBDA = 1e-5  # L2 regularization strength
    CLIP_VALUE = 1.0  # Gradient clipping threshold

    # Initialize models
    simple_rnn = SimpleRNN(len(vocab), EMBEDDING_DIM, HIDDEN_SIZE, OUTPUT_SIZE, DROPOUT_RATE).to(device)
    packed_rnn = PackedRNN(len(vocab), EMBEDDING_DIM, HIDDEN_SIZE, OUTPUT_SIZE, DROPOUT_RATE).to(device)
    bidirectional_rnn = BidirectionalRNN(len(vocab), EMBEDDING_DIM, HIDDEN_SIZE, OUTPUT_SIZE, DROPOUT_RATE).to(device)
    gru_model = GRUModel(len(vocab), EMBEDDING_DIM, HIDDEN_SIZE, OUTPUT_SIZE, N_LAYERS, DROPOUT_RATE).to(device)

    # Dictionaries to store results
    accuracies = {"Simple RNN": [], "Packed RNN": [], "Bidirectional RNN": [], "GRU": []}
    times = {"Simple RNN": 0, "Packed RNN": 0, "Bidirectional RNN": 0, "GRU": 0}
    learning_rates = {"Simple RNN": [], "Packed RNN": [], "Bidirectional RNN": [], "GRU": []}

    # Training loop
    for model_name, model in [("Simple RNN", simple_rnn), ("Packed RNN", packed_rnn), ("Bidirectional RNN", bidirectional_rnn), ("GRU", gru_model)]:
        print(f"\nTraining {model_name}")
        optimizer = Adam(model.parameters(), lr=LEARNING_RATE, weight_decay=L2_LAMBDA)
        scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=1, verbose=True)
        criterion = nn.CrossEntropyLoss()

        start_time = time.time()
        for epoch in range(EPOCHS):
            train_loss = train(model, train_dataloader, optimizer, criterion, device, L2_LAMBDA, CLIP_VALUE)
            test_loss, test_acc = evaluate(model, test_dataloader, criterion, device)
            accuracies[model_name].append(test_acc)
            learning_rates[model_name].append(optimizer.param_groups[0]['lr'])

            print(f"Epoch {epoch+1}/{EPOCHS}, Train Loss: {train_loss:.4f}, Test Loss: {test_loss:.4f}, Test Acc: {test_acc:.4f}, LR: {optimizer.param_groups[0]['lr']:.6f}")

            # Step the scheduler
            scheduler.step(test_loss)

        end_time = time.time()
        times[model_name] = end_time - start_time
        print(f"Total training time for {model_name}: {times[model_name]:.2f} seconds")

    # Visualize results
    visualize_results(accuracies, times, learning_rates)
```
